## Load the extension for Rmagic

Former extension `rmagic` is part of [rpy2](http://rpy.sourceforge.net/rpy2.html) as of [rpy2 2.4](http://rpy.sourceforge.net/rpy2/doc-2.4/html/index.html).  Many online tutorials still use the old way of loading rmagic (`%load_ext rmagic`), but [the current correct method](http://rpy.sourceforge.net/rpy2/doc-2.4/html/interactive.html#module-rpy2.ipython.rmagic) is used below (`%load_ext rpy2.ipython`).  The functionality is still referred to in this document as Rmagic because, well, it's snazzier than "rpy2 IPython magic integration" and it's more easily searched.

````python
%load_ext rpy2.ipython
````
